### 10. Web Scraping Basics (requests, Beautiful Soup)

```python
# import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://quotes.toscrape.com/"
response = requests.get(url)

print("Status Code:", response.status_code)
print("Preview of HTML:\n")
print(response.text[500:1500])
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")

print("Page title:", soup.title)
print("Title text:", soup.title.text)
```

```python
quotes = soup.find_all("span",class_="text")
authors = soup.find_all("small",class_="author")

for q, a in zip (quotes,authors):
    print(f"{q.text} - {a.text}"  )
```

```python
link = [a['href'] for a in soup.find_all("a", href=True)]
print("First 11 links", link[:11])
```

```python
all_q = []
for page in range (1,4):
    url = f"https://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    quotes = soup.find_all("span",class_ ="text")
    authors = soup.find_all("small",class_="author")
    for q, a in zip (quotes, authors ):
        all_q.append((q.text, a.text))
        
print("Total quotes scraped", len(all_q))
all_q[:5]
```

```python
df = pd.DataFrame(all_q, columns=["Quotes", "Authors"])
# df.to_csv("quotes.csv",index=False)
df.head()
```

### 12. Introduction to Data Science Workflow

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Alan","Bob", "Charles", "Dave","Even", "Frankly", "Gunn","Heller", "Ivan", "John"],
    "Hrs_st": [2, 3, 4, 5, 6, 8, 7, 9, 5, 10],
    "Score": [32, 44, 56, 64, 71,84, 88, None, 92, 97]
}
df = pd.DataFrame(data)
# df.to_csv("students.csv", index=False)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Score'] = df['Score'].fillna(df['Score'].mean())
df.rename(columns={"Hrs_st": "Hours"},inplace = True)
df["Pass"] = df["Score"].apply(lambda x: "Pass" if x<=50 else "Fail")
df.head()
```

```python
#Analisis y visualizacion
df.describe()
```

```python
print("data types antes:", df.dtypes)
df_clean = df[pd.to_numeric(df['Hours'], errors="coerce").notnull() & 
              pd.to_numeric(df['Score'], errors="coerce").notnull()]

print("data types despues:", df_clean.shape)

plt.scatter(df['Hours'], df['Score'])
plt.xlabel("Hours studied")
plt.ylabel("Score")
plt.title("Hours vs Score")
plt.show()

numeric_df = df_clean.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x = df[['Hours']]
y = df[['Score']]

model = LinearRegression()
model.fit(x,y)
prediction = model.predict([[5]])
print("Prediccion para 5 horas", prediction[0])
```

### 14. Pandas Basics (data frames, filtering, grouping)

```python
import pandas as pd
data = {
    "Name": ["Alan","Bob", "Charles", "Dave","Even", "Frankly", "Gunn","Heller", "Ivan", "John"],
    "age": [20, 32, 41, 15, 26, 28, 47, 49, 35, 10],
    "Score": [32, 44, 56, 64, 71,84, 88, 10, 92, 97]
}
df = pd.DataFrame(data)
# df.to_csv("students.csv", index=False)
df.head()
print(df[(df['age'] <45) & (df["Score"] > 64)])
print(df[(df['age'] <45) | (df["Score"] > 64)])
```

```python
df ['department'] =["IT","RH", "HYS","HYS","RH","Finance","RH", "HYS","HYS","RH"]
print(df.groupby("department")['Score'].mean())
print(df.groupby("department")['Score'].count())
```

```python
agg_data = df.groupby("department").agg({
    'Score': ["mean","max", "min"],
    "age": "mean"
    
})
print(agg_data)
```
