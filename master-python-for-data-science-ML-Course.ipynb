```python
for i in range (0,5):
    print(i)
```

```python
total = 0
for i in range (1,12):
    total += i

print("Suma: ", total)
    
```

```python
#Nested loop
for i in range (1,5):
    for j in range (1,5):
        print( i, "*", j, "=", i*j)
```

```python
# While loop
#Se repite hasta que sea falso 

count = 1
while count <=12:
    print("Count is: ", count)
    count +=1
```

```python
# Break dentro while loop

j = 1
while True:
    print(j)
    if j == 6:
        break
    
    j+= 1
```

```python
def add (a, b):
    return a + b
resultado = add (15, 10)
print(resultado)
```

```python
def pow(base, exponent =2):
    return base ** exponent

resultado = pow(6)
print(resultado)
```

```python
# Retorna varios valores

def cal (a, b):
    return a+b, a-b, a*b

suma, dif, producto = cal(10, 8)

print("Suma:", suma)
print("Diferencia:", dif)
print("Producto:", producto)
```

```python
def wor_information (name, age, no):
    print(f"Name: {name}, Age: {age}, Number: {no}")
    
wor_information ("Jose", 32, 12348)
```

```python
def total(*number): #no le digo cuantos valores voy a agrgar
    return sum(number)

print(total(5,10,15))
print(total(10,20))
```

```python
# Diferentes detalles
def print_info(**detail):
    for key, value in detail.items():
        print(f"{key}: {value}")

print_info (name="John", age=34, state="Birmingham", id_no = 1234567)
```

```python
 #Lambda function
def sqo(x):
    return x*x
#Es lo mismo pero en una sola linea
sqo_lambda = lambda x: x*x
print(sqo(8))
print(sqo_lambda(8))
```

```python
# set: colección desordenada de elementos únicos e inmutables, como números, cadenas de texto o tuplas

frutas = {"Banana", "Naranja","Peras"}
print(frutas)

frutas = {'Banana', 'Naranja','Peras'}
print(frutas)

numbers = {1, 1,2 , 3 , 4 , 5 , 6, 7, 8, 8, 9 , 9}
print(numbers) #No repite los numeros , porque?, es un set de datos

#Set vacio
set_vacio = set()
print(type(set_vacio))
```

```python
frutas.add('kiwi') 
print(frutas) #lo agrega 2dp
frutas.remove('Banana')
print(frutas)
```

```python
#Vemos si esta o no
print("kiwi" in frutas)
```

```python
set1 = {1, 2, 3}
set2 = {3, 4 ,5}

print("Union:", set1 | set2)
print("Intersect:", set1 & set2)
# La diferencia de conjuntos (-) devuelve todos los elementos que están en set1 pero NO están en set2.
print("Diferences:", set1 - set2)
# La diferencia de conjuntos (-) devuelve todos los elementos que no están en set1 ni en el set2.
print("Diferences:", set1 ^ set2)
```

```python
no = [1, 2, 2, 2, 3, 4, 5, 5]
unique_no = set(no) #Se eliminaron los duplicados de la lista
print(unique_no)
```

```python
# frozenset es igual que un set, pero inmutable (no se puede modificar).
fs = frozenset([1, 2, 3, 4, 5])
print(fs)
```

```python
import csv
data = [
    ["Name", "Age", "City"], 
    ["Eren", 32, "NY"],
    ["Lucianito", 34, "LA"],
    ["Jose", 31, "Cor"],
    
]
print(data)
with open("info.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
with open("info.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
   
import csv

with open("info_dict.csv", "w", newline="") as f:
    fieldnames = ["Name", "Age", "City"] 
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"Name": "Diab", "Age": 23, "City": "Berlin"})

```

```python
with open("info_dict.csv", "r") as f:
    reader =  csv.DictReader(f)
    for row in reader:
        print(row["Name"], row["City"])
```
